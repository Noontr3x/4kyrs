```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

data = pd.read_csv('flight_durations.csv', encoding='utf-8')

durations = data['duration'].values.reshape(-1, 1)

n_clusters = 3  # Можно изменить количество кластеров по необходимости

kmeans = KMeans(n_clusters=n_clusters, random_state=42)
data['cluster'] = kmeans.fit_predict(durations)

linguistic_labels = ["Korotkiy perelet", "Sredniy perelet", "Dlinnyy perelet"]

data['label'] = data['cluster'].apply(lambda x: linguistic_labels[x])

print(data)

data.to_csv('clustered_flight_durations.csv', index=False, encoding='utf-8')

plt.scatter(data['duration'], np.zeros_like(data['duration']), c=data['cluster'])
plt.xlabel("Dlitel'nost' pereleta (minuty)")
plt.title("Klasterizatsiya dlitel'nosti pereletov")
plt.show()
```
