```python
import numpy as np
import matplotlib.pyplot as plt

class TrapezoidalFuzzySet:
    def __init__(self, a, b, c, d):
        self.a = a
        self.b = b
        self.c = c
        self.d = d

    def membership(self, x):
        if x <= self.a:
            return 0.0
        elif self.a < x <= self.b:
            return (x - self.a) / (self.b - self.a)
        elif self.b < x <= self.c:
            return 1.0
        elif self.c < x <= self.d:
            return (self.d - x) / (self.d - self.c)
        else:
            return 0.0

def union_fuzzy_sets(fuzzy_set1, fuzzy_set2, x_values):
    union_membership = np.zeros_like(x_values)
    for i, x in enumerate(x_values):
        union_membership[i] = max(fuzzy_set1.membership(x), fuzzy_set2.membership(x))
    return union_membership

def main():
    #    
    fuzzy_set1 = TrapezoidalFuzzySet(a=1, b=3, c=5, d=7)
    fuzzy_set2 = TrapezoidalFuzzySet(a=4, b=5, c=7, d=9)
    
    #   
    x_values = np.linspace(0, 10, 100)
    
    #    
    union_membership = union_fuzzy_sets(fuzzy_set1, fuzzy_set2, x_values)
    
    membership_fuzzy_set1 = [fuzzy_set1.membership(x) for x in x_values]
    membership_fuzzy_set2 = [fuzzy_set2.membership(x) for x in x_values]
    
    plt.plot(x_values, membership_fuzzy_set1, label='  1')
    plt.plot(x_values, membership_fuzzy_set2, label='  2')
    plt.plot(x_values, union_membership, label='  ', linestyle='--')
    plt.xlabel('x')
    plt.ylabel(' ')
    plt.legend()
    plt.title('      ')
    plt.show()

if __name__ == '__main__':
    main()
```
